***Introduction to Dictionary***

---

Dictionary allows user to write the data in the form of keys and values.
Dictionaries are enclosed inside curly brackets {}.
Keys and Values are separated by colon
Every key value pair is separated by a coma(,).

```python
employee_data = {
    "name":"Ashu",
    "age":12,
    "gender":"male"
}
```

```python
print(employee_data)
```

```python
print(employee_data["name"])
```

***Iteration in Dictionary***

---



```python
for i in employee_data:
  print(i)
```

```python
for i in employee_data:
  print(employee_data[i])
```

```python
#using value Function
for i in employee_data.values():
  print(i)
```

```python
#using items Function
for x,y in employee_data.items():
  print(x,"-",y)
```

***Dictionary Functions***

---



```python
#get
#item
#keys
#values
#copy
```

```python
employee_data = {
    "name":"Ashu",
    "age":12,
    "gender":"male",
    "doj":"22/01/2019"

}
```

```python
print(employee_data.get("doj"))
```

```python
x = employee_data.items()
print(x)
```

```python
x = employee_data.keys()
print(x)
```

```python
x = employee_data.values()
print(x)
```

```python
d = employee_data.copy()
print(d)
```

```python
#setdefault - The setdefault() method returns the value of the key if it's in the dictionary. If the key is not found, it inserts the key with the specified default value and returns that value. If no default value is given, it defaults to None.
#update - The update() method adds key-value pairs from another dictionary (or an iterable of key-value pairs) into the current dictionary. If a key already exists, its value is overwritten.
#pop - The pop() method removes the specified key and returns its value. If the key is not found, it raises a KeyError unless a default value is provided as the second argument.
#popitem - The popitem() method removes and returns the last inserted key-value pair from the dictionary as a tuple ((key, value)). It raises a KeyError if the dictionary is empty.
#clear - The clear() method removes all items from the dictionary, making it empty.
```

```python
employee_data = {
    "name":"Ashu",
    "age":12,
    "gender":"male",
    "doj":"22/01/2019"

}
```

```python
employee_data.setdefault("age", 18) #unchanged
employee_data.setdefault("salary", 50000) #added
print(employee_data)
```

```python
employee_data.update({"age": 18, "city": "Delhi"}) #age is updated to 18, and city is added.
print(employee_data)
```

```python
employee_data.pop("doj")
print(employee_data)
```

```python
employee_data.popitem() #The pair is removed and returned as a tuple.
```

```python
employee_data.clear()
print(employee_data)
```

***Nested Dictonaries***

```python
print(employee)
```

```python
print(employee[1]["age"])
print(employee[2]["gender"])
```
