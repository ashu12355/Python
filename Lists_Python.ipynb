***List Comprehension***

---



```python
l1 = [30 ,40,50,60]
l2 = []
for i in l1:
  if i > 30:
    l2.append(i)
print(l2)
print(l1)
```

```python
l3 = [i for i in l1] #List Comprehension
print(l3)
```

```python
l3 = [i for i in l1 if i > 30] #List Comprehension
print(l3)
```
